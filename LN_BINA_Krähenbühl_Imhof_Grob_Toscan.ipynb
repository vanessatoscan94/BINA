Beim Erwerb von Renditeimmobilien spielt der Standort, bestehend aus Mikro- und Makrolage, eine essenzielle Rolle, da er massgebend für die zukünftige Wertentwicklung sein wird.

Die Begriffe Mikro- und Makrolage werden häufig im Bereich der Immobilienbewertung und -beschreibung verwendet.

Die Mikrolage bezieht sich auf die unmittelbare Umgebung eines bestimmten Standorts. Hier werden Faktoren wie die Nähe zu Einkaufsmöglichkeiten, öffentlichen Verkehrsmitteln, Schulen und Arbeitsplätzen sowie die Qualität der Infrastruktur, der Nachbarschaft und der Landschaft berücksichtigt.

Die Makrolage hingegen bezieht sich auf die Lage eines Objekts im grösseren Kontext. Hier werden Faktoren wie die Region, die Stadt oder der Stadtteil, in dem sich das Objekt befindet, berücksichtigt. Auch politische, wirtschaftliche und gesellschaftliche Faktoren können hier eine Rolle spielen, beispielsweise die Arbeitsmarktsituation oder die Entwicklung des regionalen Immobilienmarktes.

Insgesamt ergibt sich die Lagebewertung eines Objekts aus der Betrachtung sowohl der Mikro- als auch der Makrolage.





### **1.1 Zielsetzung**
Das Makro- Mikrolagen- Datentool soll den Usern schnell und einfach die Möglichkeit bieten, attraktive Gemeinden für eine Investition in Renditeimmobilien identifizieren zu können. Es soll helfen, die bestmögliche Investitionsentscheidung zu treffen, indem es relevante Informationen über alle Schweizer Gemeinden und deren Aussichten für zukünftige Wertentwicklung aufzeigt und Risikofaktoren identifiziert, um passende Entscheidungsgrundlagen zu definieren.

Folgende Informationsgrundlagen müssen bei einer Investition in Renditeimmobilien in Betracht gezogen werden:


*   Standort: Die Lage einer Immobilie ist ein wichtiger Faktor, der die Rendite beeinflussen kann. Primär geht es hier um Faktoren wie Infrastruktur, Anbindung an den öffentlichen Nahverkehr, Verfügbarkeit von Schulen und Einkaufsmöglichkeiten, welche Standorte wesentlich aufwerten können.
*   Markt- Lage und Entwicklung: Auch die generelle Entwicklung der Marktgegebenheiten mit der allgemeinen Wirtschaftslage, Zinssätze und andere wirtschaftliche Trends können einen wesentlichen Einfluss auf die Rendite haben und müssen entsprechend berücksichtigt werden.
*   Leerstandsziffer: Die Leerstandsziffer gibt an, wie viele Einheiten einer Immobilie im Leerstand sind. Eine hohe Leerstandsquote kann einen Standort wesentlich negativ beeinflussen.
*   Alter und Zustand der Immobilie: Diese Faktoren können die Rendite einer Immobilie wesentlich beeinflussen, da ältere Immobilien oft höhere Instandhaltungskosten als neuere Immobilien verursachen und mit zeitnahen kapitalintensiven Renovationskosten zu rechnen sind.
*   Mieteinnahmen: Die Höhe der Mieteinnahmen ist ein wichtiger Faktor, der die Rendite beeinflussen kann. Es muss festgestellt werden, ob sich diese im Marktniveau befinden oder ggf. noch optimiert werden können.
*   Steuern und andere Kosten: In der föderalistischen Schweiz gibt es wesentliche Unterschiede bei der steuerlichen Handhabung.

```python
# Auf Excel-Datei zugreifen
file_path = '/content/gdrive/MyDrive/datasets/table_near_centers.xlsx'
data = pd.read_excel(file_path)
# Warnung ignorieren
warnings.filterwarnings("ignore", category=UserWarning, module="openpyxl")
```

```python
# Eindeutige Werte in der Spalte "city"
unique_cities = data['city'].unique()

# Breite der Balken
bar_width = 0.35

# Für jeden eindeutigen Wert ein separates Balkendiagramm erstellen
for city in unique_cities[:5]:  # Begrenzung der Schleife auf die ersten fünf Gemeinden
    # Filtern der Daten für den aktuellen Wert in der Spalte "city"
    city_data = data[data['city'] == city]

    # Eindeutige Werte in der Spalte "near_center"
    unique_near_center = city_data['near_center'].unique()

    # Positionen für die X-Achse
    x = np.arange(len(unique_near_center))

    # Balkendiagramm erstellen
    for i, near_center_val in enumerate(unique_near_center):
        # Filtern der Daten für den aktuellen Wert in der Spalte "near_center"
        near_center_data = city_data[city_data['near_center'] == near_center_val]

        # Werte für "by=car"
        by_car_values = near_center_data[near_center_data['by'] == 'car']['distance']

        # Werte für "by=transport"
        by_transport_values = near_center_data[near_center_data['by'] == 'transport']['distance']

        # Balken für "by=car" erstellen
        plt.bar(x[i] - bar_width / 2, by_car_values, width=bar_width, label='by=car', color='gray')

        # Balken für "by=transport" erstellen
        plt.bar(x[i] + bar_width / 2, by_transport_values, width=bar_width, label='by=transport', color='green')

        # Wert oberhalb des Balkens hinzufügen
        for j, val in enumerate(by_car_values):
            plt.text(x[i] - bar_width / 2, val + 1, str(val), ha='center', va='bottom', color='black')

        for j, val in enumerate(by_transport_values):
            plt.text(x[i] + bar_width / 2, val + 1, str(val), ha='center', va='bottom', color='black')

    # Achsenbeschriftungen und Titel hinzufügen
    plt.xlabel('Gemeinden in der Nähe')
    plt.ylabel('Distanz in Minuten')
    plt.title(f'Distanzen zu anderen Gemeinden für {city}')

    # X-Achse beschriften
    plt.xticks(x, unique_near_center)

    # Y-Achse anpassen
    plt.ylim(0, city_data['distance'].max() + 10)
    plt.yticks(np.arange(0, city_data['distance'].max() + 10, 10))

    # Legende anzeigen
    legend_elements = [
        plt.Line2D([0], [0], color='gray', lw=4, label='Distanz mit dem Auto'),
        plt.Line2D([0], [0], color='green', lw=4, label='Distanz mit dem ÖV')
    ]
    plt.legend(handles=legend_elements)

    # Diagramm anzeigen
    plt.show()
```

### **Code Auswertung Attraktivität der Gemeinden**

```python
# Auf Excel-Datei zugreifen
file_path_1 = '/content/gdrive/MyDrive/datasets/table_residents_new.xlsx'
file_path_2 = '/content/gdrive/MyDrive/datasets/table_tax.xlsx'

# Einlesen der Daten
table1 = pd.read_excel(file_path_1)
table2 = pd.read_excel(file_path_2)

# Zusammenführen der Tabellen basierend auf dem Schlüssel "city"
merged_table = pd.merge(table1, table2, on='city', how='inner')
```

```python
# Überprüfen der Datenstruktur
merged_table.info()
```

```python
# Erste Einträge anzeigen
merged_table.head()
```

```python
# Filtern der Daten
filtered_data = merged_table.loc[merged_table['type'] == 'Gemeinde'].copy()

# Konvertieren der Spalte "Steuerfuss" in numerischen Datentyp
filtered_data['Steuerfuss'] = pd.to_numeric(filtered_data['Steuerfuss'], errors='coerce')

# Bereinigen der Spalte "Einkommen" und Konvertieren in numerischen Datentyp
filtered_data['Einkommen'] = filtered_data['Einkommen'].str.replace("'", "").astype(float)

# Einteilung in die Kategorien A-Gemeinden, B-Gemeinden und C-Gemeinden basierend auf Einkommen, Steuerfuss und Entwicklung
filtered_data['Kategorie'] = 'C-Gemeinden'  # Alle Gemeinden zunächst der Kategorie "C-Gemeinden" zuweisen

# Bestimmung des Medians für die Variablen "Einkommen" und "Steuerfuss"
median_income = filtered_data['Einkommen'].median()
median_tax = filtered_data['Steuerfuss'].median()

# Einteilung der Gemeinden in die Kategorien A-Gemeinden, B-Gemeinden und C-Gemeinden basierend auf den Medianen
filtered_data.loc[(filtered_data['Einkommen'] > median_income) & (filtered_data['Steuerfuss'] < median_tax) & (filtered_data['Entwicklung'] == 1), 'Kategorie'] = 'A-Gemeinden'
filtered_data.loc[(filtered_data['Einkommen'] < median_income * 0.9) & (filtered_data['Steuerfuss'] > median_tax * 1.1) & (filtered_data['Entwicklung'].isin([0, 1])), 'Kategorie'] = 'B-Gemeinden'

# Festlegen der Reihenfolge der Kategorien
category_order = ['A-Gemeinden', 'B-Gemeinden', 'C-Gemeinden']
filtered_data['Kategorie'] = pd.Categorical(filtered_data['Kategorie'], categories=category_order, ordered=True)

# Sortieren der Daten nach der Kategorie
filtered_data.sort_values('Kategorie', inplace=True)

# Erstellen der Treemap
fig = px.treemap(filtered_data, path=['Kategorie', 'city'], values='Einkommen',
                 color='Kategorie', color_discrete_sequence=['#1f77b4', '#ff7f0e', '#2ca02c'],
                 hover_data={'Steuerfuss': ':.2f', 'Einkommen': ':.2f', 'Entwicklung': True},
                 labels={'Einkommen': 'Einkommen in CHF', 'Steuerfuss': 'Steuerfuss'})

# Anpassen der Legende
fig.update_layout(legend=dict(title=dict(text='Kategorie')))

# Hinzufügen des Titels
fig.update_layout(title='Gemeindeattraktivität')

# Anzeigen der Treemap
fig.show()
```

Leere Zellen werden mit NaN-Werten aufgefüllt, um sie später handhaben zu können. Die Spalten "Einkommen" und "Steuerfuss" werden in numerische Werte umgewandelt. Zeilen mit fehlenden Werten in den Spalten "Einkommen" und "Steuerfuss" werden entfernt und die Daten werden nach dem Einkommen sortiert. Danach wird der Korrelationskoeffizient zwischen Einkommen und Steuerfuss berechnet.

Ein Scatterplot wird erstellt, wobei das Einkommen auf der x-Achse, der Steuerfuss auf der y-Achse und die Städte als Hover-Information dargestellt werden. Eine Trendlinie wird hinzugefügt, um den allgemeinen Trend darzustellen. Eine Korrelationslinie wird als separater Trace zum Plot hinzugefügt, indem die x-Werte des Einkommens und die y-Werte des Einkommens multipliziert mit dem Korrelationskoeffizienten verwendet werden.

Die Achsenbeschriftungen und die Skalierung des Plots werden angepasst, und ein Titel wird hinzugefügt und der Plot danach angezeigt.


### **Code Korrelation zwischen dem Steuerfuss und dem Einkommen**

```python
# Auf Excel-Datei zugreifen
file_path_1 = '/content/gdrive/MyDrive/datasets/table_residents_new.xlsx'
file_path_2 = '/content/gdrive/MyDrive/datasets/table_tax.xlsx'

# Einlesen der Daten
table1 = pd.read_excel(file_path_1)
table2 = pd.read_excel(file_path_2)

# Zusammenführen der Tabellen basierend auf dem Schlüssel "city"
merged_table = pd.merge(table1, table2, on='city', how='inner')
```

```python
# Filtern der Daten basierend auf dem Wert in der Spalte "type"
filtered_table = merged_table[merged_table['type'] == 'Gemeinde'].copy()

# Entfernen von Apostrophen, Kommas und negativem Vorzeichen aus der Spalte "Einkommen"
filtered_table['Einkommen'] = filtered_table['Einkommen'].str.replace("'", "").str.replace(",", "").str.replace("-", "")

# Entfernen von negativem Vorzeichen aus der Spalte "Steuerfuss"
filtered_table['Steuerfuss'] = filtered_table['Steuerfuss'].str.replace("-", "")

# Füllen von leeren Zellen mit NaN-Werten
filtered_table['Einkommen'] = pd.to_numeric(filtered_table['Einkommen'], errors='coerce')
filtered_table['Steuerfuss'] = pd.to_numeric(filtered_table['Steuerfuss'], errors='coerce')

# Entfernen von Zeilen mit fehlenden Werten in den Spalten "Einkommen" und "Steuerfuss"
filtered_table = filtered_table.dropna(subset=['Einkommen', 'Steuerfuss'])

# Sortieren der Datenpunkte nach Einkommen
filtered_table = filtered_table.sort_values(by='Einkommen')

# Berechnen der Korrelationskoeffizienten
correlation_coefficient = filtered_table['Einkommen'].corr(filtered_table['Steuerfuss'])

# Erstellen des Scatterplots mit vollständigen Datenpunkten und City als Hover-Information
fig = px.scatter(filtered_table, x='Einkommen', y='Steuerfuss', trendline="ols", trendline_color_override="red",
                 hover_data=['city'])

# Hinzufügen der Korrelationslinie als separaten Trace
fig.add_trace(go.Scatter(
    x=filtered_table['Einkommen'],
    y=filtered_table['Einkommen'] * correlation_coefficient,
    mode='lines',
    name='Korrelationslinie',
    line=dict(color='red', width=2)
))

# Anpassen der Achsenbeschriftungen und Skalierung
fig.update_layout(
    xaxis=dict(title='Einkommen'),
    yaxis=dict(title='Steuerfuss', range=[0, filtered_table['Steuerfuss'].max() + 1]),
    title='Korrelation zwischen Einkommen und Steuerfuss'
)

# Anzeigen des Plots
fig.show()
```

**Brechnung Korrelationskoeffizient**

```python
# Berechnen des Korrelationskoeffizienten
correlation_coefficient = filtered_table['Einkommen'].corr(filtered_table['Steuerfuss'])

# Ausgabe des Korrelationskoeffizienten
print("Korrelationskoeffizient:", correlation_coefficient)
```

Ein Korrelationskoeffizient von -0.2520276705978861 bedeutet, dass zwischen den beiden Variablen "Einkommen" und "Steuerfuss" eine schwache negative Korrelation besteht.

Die negative Vorzeichen des Korrelationskoeffizienten (-0.2520276705978861) zeigt an, dass es eine inverse Beziehung zwischen den beiden Variablen gibt. Das bedeutet, dass höhere Werte der einen Variable mit niedrigeren Werten der anderen Variable korrelieren.

Der Plot visualisiert die Beziehung zwischen Einkommen und Steuerfuss für die ausgewählten Gemeinden und zeigt auch die Gemeinden als Hover-Information an. Die Trendlinie und die Korrelationslinie geben eine visuelle Darstellung der Korrelation zwischen Einkommen und Steuerfuss. Es ist zu erkennen, dass Gemeinden mit tiefem Steuerfuss tendenziell auch über ein höheres Einkommen verfügen, da sie wohlhabende Personen anziehen. Besonders die Gemeinde Vaux sur Morges sticht dabei besonders heraus. Dies hat die Waadtländer Gemeinde einer einzigen Person zu verdanken: André Hoffmann, Vizepräsident des Pharmakonzerns Roche.
